```python
req_amazon = amazon_data[["product_name",'retail_price','discounted_price']]
req_flipkart = flipkart_data[["product_name",'retail_price','discounted_price']]

df2=req_amazon.compare(req_flipkart,keep_shape=True, keep_equal=True)

```

```python
df2.rename(columns={"self":"Product Name Amazon"})

product_name=df2["product_name"].rename(columns={"self":"Product Name in Amazon","other":"Product Name in Flipkart"})
discounted=df2["discounted_price"].rename(columns={"self":"Discounted Price in Amazon","other":"Discounted Price in Flipkart"})
retail=df2["retail_price"].rename(columns={"self":"Retail Price in Amazon","other":"Retail Price in Flipkart"})

solution =  pd.concat([product_name,retail,discounted.reset_index(drop=True)],ignore_index=True)
solution.head()
# df2.head()
# retail.head()
# discounted.head()
```
